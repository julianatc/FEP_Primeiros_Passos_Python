#Calcular IMC

```python
#Criei uma funçāo para calcular o IMC (Indice de massa corpórea)

def calcula_IMC (peso, altura):
  IMC = peso / (altura ** 2) #IMC = peso (em Kg) dividido pela altura ao quadrado (em metros)

  if IMC >= 18.5 and IMC <= 24.9:
    tipo = 'normal'
  elif IMC >= 25.0 and IMC <= 29.9:
    tipo = 'sobrepeso'
  elif IMC > 30.0:
    tipo = 'obesidade'
  elif IMC > 40.0:
    tipo = 'obesidade grave'
  else:
    tipo = 'abaixo do peso'

  return IMC, tipo

  #print(f'IMC = {round(IMC,2)}')
  #print(f'Classificaçāo do IMC: {tipo}' )


```

```python
# testando a funçāo
imc, classificacao = calcula_IMC(61.1, 1.71)
print(f'IMC = {round(imc,2)}')
print(f'Classificaçāo do IMC: {classificacao}')
```

# Entrada de dados

```python
nome = input('Informe o nome do aluno(a): ')
idade = input('Informe a idade do aluno(a): ')
peso = input('Informe o peso (Kg) do aluno(a): ')
altura = input('Informe a altura (m) do aluno(a):')
cpf = input('Informe o CPF do aluno(a): ')

#chamada da funcao para calcular IMC 
imc, classificacao = calcula_IMC(float(peso),float(altura))

#Imprimindo dados
print('CPF do aluno:',cpf)
print('Nome do aluno:', nome)
print('Idade do aluno:', idade)
print('Peso do aluno:', peso)
print('Altura do aluno: ', altura)
print(f'IMC = {round(imc,2)}')
print(f'Classificaçāo do IMC: {classificacao}')
```
